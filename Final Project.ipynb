### Remove everything but letters, periods, and spaces

```python
article_text = re.sub(r'[^A-Za-z. ]', '', article_text)
```

### Create character trigram model

```python
ngrams = {}
chars = 3

for i in range(len(article_text)-chars):
    seq = article_text[i:i+chars]
    print(seq)
    if seq not in ngrams.keys():
        ngrams[seq] = []
    ngrams[seq].append(article_text[i+chars])
```

```python
curr_sequence = article_text[0:chars]
output = curr_sequence
for i in range(200):
    if curr_sequence not in ngrams.keys():
        break
    possible_chars = ngrams[curr_sequence]
    next_char = possible_chars[random.randrange(len(possible_chars))]
    output += next_char
    curr_sequence = output[len(output)-chars:len(output)]

print(output)
```

### Create word trigram model

```python
ngrams = {}
words = 3

sent_tokens = sent_tokenize(article_text)

padded_tokens = []

for sent in sent_tokens:
#     sent = re.sub(r'[.]', ' </s> ', sent)
#     padded_tokens.append("<s> <s> " + sent)
    sent = re.sub(r'[.]', ' EOS ', sent)
    padded_tokens.append(" BOS BOS BOS " + sent)
    
padded_tokens = " ".join(padded_tokens)

#words_tokens = nltk.word_tokenize(padded_tokens)
words_tokens = tk.tokenize(padded_tokens)
for i in range(len(words_tokens)-words):
    seq = ' '.join(words_tokens[i:i+words])
    print(seq)
    if seq not in ngrams.keys():
        ngrams[seq] = []
    ngrams[seq].append(words_tokens[i+words])
```

### Generate sentence based on randomly selecting trigrams

```python
def generate_sentence(start):
    start = re.sub(r'[^A-Za-z. ]', '', start)
    start = tk.tokenize(start.lower())
    curr_sequence = ' '.join(start[len(start) - words:len(start)])
    print(curr_sequence)
    output = curr_sequence
    for i in range(50):
        if curr_sequence not in ngrams.keys():
            break
        possible_words = ngrams[curr_sequence]
        next_word = possible_words[random.randrange(len(possible_words))]
        output += ' ' + next_word
        seq_words = tk.tokenize(output)
        curr_sequence = ' '.join(seq_words[len(seq_words) - words:len(seq_words)])
    return output
        
print(generate_sentence("computer science departments"))
```

### Count instances of each trigram

```python
ngrams_counts = {}

for ngram in ngrams:
    count = dict(Counter(ngrams[ngram]))
    ngrams_counts[ngram] = count
```

```python
ngrams_counts
```

### Calculate probabilities for each trigram

```python
probabilities = {}

for key in ngrams_counts.keys():
    
    temp = {}
    
    s = sum(ngrams_counts[key].values())
    
    for k in ngrams_counts[key].keys():
                
        temp[k] = ngrams_counts[key][k]/s
#         ngrams_counts[key][k] = ngrams_counts[key][k]/s
    
    probabilities[key] = []
    
#     probabilities[key].append(ngrams_counts[key])
    probabilities[key].append(temp)
    
```

### Generate Sentence using max probability for each N-gram

```python
def generate_sentence_probabilities(start):
    sentence_prob = 1
    start = re.sub(r'[^A-Za-z. ]', '', start)
    start = tk.tokenize(start.lower())
    curr_sequence = ' '.join(start[len(start) - words:len(start)])
    print(curr_sequence)
    output = curr_sequence
    for i in range(50):
        if curr_sequence not in ngrams.keys():
            break
        possible_words = probabilities[curr_sequence][0]
        next_word = max(possible_words, key=possible_words.get)
        sentence_prob = sentence_prob * possible_words[next_word]
#         print(possible_words)
#         print(next_word)
        output += ' ' + next_word
#         print(output)
        seq_words = tk.tokenize(output)
        curr_sequence = ' '.join(seq_words[len(seq_words) - words:len(seq_words)])
    return output, sentence_prob

out = generate_sentence_probabilities('Theoretical Computer Science')
print(out)
```

### Calculate Perplexity

```python
def calc_perplexity(prob, length):
    p = np.log2(prob)/length
    PP = np.power(2, -p)
    return PP

PP = calc_perplexity(out[1], 50)
print(PP)
```

### Add one smoothing

```python
def add_one(ngram_counts):
    output = {}
    for ngram in ngram_counts:
        temp = {}
        for ng in ngram_counts[ngram]:
            temp[ng] = ngram_counts[ngram][ng] + 1
        output[ngram] = []
        output[ngram].append(temp)
    return output

add_one_ngram_counts = add_one(ngrams_counts)
```

```python
add_one_ngram_counts
```

### Calculate new probabilities for add one smoothing

```python
add_one_probabilities = {}

for key in add_one_ngram_counts.keys():
    
    temp = {}
    
    s = sum(add_one_ngram_counts[key][0].values())
    
    for k in add_one_ngram_counts[key][0].keys():
                
        temp[k] = add_one_ngram_counts[key][0][k]/s
#         ngrams_counts[key][k] = ngrams_counts[key][k]/s
    
    add_one_probabilities[key] = []
    
#     probabilities[key].append(ngrams_counts[key])
    add_one_probabilities[key].append(temp)
    
```

### Generate Sentence using Add one smoothing

```python
def generate_sentence_probabilities_add_one(start, length):
    sentence_prob = 1
    start = re.sub(r'[^A-Za-z. ]', '', start)
    start = tk.tokenize(start.lower())
    curr_sequence = ' '.join(start[len(start) - words:len(start)])
    #print(curr_sequence)
    output = curr_sequence
    for i in range(length):
        if curr_sequence not in ngrams.keys():
            break
        possible_words = add_one_probabilities[curr_sequence][0]
        next_word = max(possible_words, key=possible_words.get)
        sentence_prob = sentence_prob * possible_words[next_word]
#         print(possible_words)
#         print(next_word)
        output += ' ' + next_word
#         print(output)
        seq_words = tk.tokenize(output)
        curr_sequence = ' '.join(seq_words[len(seq_words) - words:len(seq_words)])
    return output, sentence_prob

out = generate_sentence_probabilities_add_one('theory of computation', 50)
print(out)
```

### Beam Search

```python
def beam_search(initial_input, length, beam_width):
    
    start = re.sub(r'[^A-Za-z. ]', '', initial_input)
    start = tk.tokenize(start.lower())
    
    beam_seq = []
    beam_prob = np.ones((beam_width,1))
    output_seq = []
    
    for i in range(beam_width):
        beam_seq.append(' '.join(start[len(start) - words:len(start)]))
        output_seq.append(beam_seq[i])
    
    for i in range(length):
        
#         print(i)
        
        possible_seq = {}
        
        for j in range(beam_width):
        
            temp = {}
        
            if beam_seq[j] not in ngrams.keys():
                break
        
            possible_words = add_one_probabilities[beam_seq[j]][0]
            for key in possible_words.keys():
                temp[key] = add_one_probabilities[beam_seq[j]][0][key] * beam_prob[j]
                possible_seq[beam_seq[j]] = temp

#         print(possible_seq)
        
        for a in range(beam_width):
            beam_prob[a] = 0
        
        temp_output_seq = [None]*beam_width
        
        for key in possible_seq.keys():
            
            #print(key)
            
            for k in possible_seq[key].keys():
                
                #print(k)
                
                index = beam_prob.argmin()
                
                beam_prob[index] = 0
                
                for l in range(beam_width):
                    temp_output_seql = []
                
                if beam_prob[index] <= possible_seq[key][k]:
                    beam_prob[index] = possible_seq[key][k]
                    index_2 = list(possible_seq).index(key)
                    temp_output_seq[index] = output_seq[index_2] + ' ' + k
                #print(temp_output_seq)
                
                #print(output_seq)
        for n in range(beam_width):
            if temp_output_seq[n] == None:
                output_seq[n] = output_seq[0]
                beam_prob[n] = beam_prob[0]
                output_seq[n] = output_seq[0]
                continue
            else:
                output_seq[n] = temp_output_seq[n]
        
#         print(output_seq)
            
        for b in range(beam_width):
            
            p = word_tokenize(output_seq[b])
            t = ' '.join(p[len(p) - words:len(p)])
                                    
            beam_seq[b] = t
            
#         print(beam_seq)
#         print(beam_prob)
#         print(output_seq)
    
    return output_seq, beam_prob
    

beam_search('a database is', 25, 3)
```
